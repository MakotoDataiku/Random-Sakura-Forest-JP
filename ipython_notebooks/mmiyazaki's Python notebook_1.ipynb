```python
# -------------------------------------------------------------------------------- NOTEBOOK-CELL: CODE
# -*- coding: utf-8 -*-
import dataiku
import pandas as pd, numpy as np
from dataiku import pandasutils as pdu
```

```python
# -------------------------------------------------------------------------------- NOTEBOOK-CELL: CODE
# Read recipe inputs
three_cities_blooming_day_joined = dataiku.Dataset("three_cities_blooming_day_joined")
df = three_cities_blooming_day_joined.get_dataframe()
```

```python
# -------------------------------------------------------------------------------- NOTEBOOK-CELL: CODE
df = df.sort_values(['place', 'date_parsed'])

# -------------------------------------------------------------------------------- NOTEBOOK-CELL: CODE
```

```python
df.head(366)
```

```python
# -------------------------------------------------------------------------------- NOTEBOOK-CELL: CODE
df.tail()
```

```python
# -------------------------------------------------------------------------------- NOTEBOOK-CELL: CODE
d = {}
for city in df.place.unique():
    d[city] = {}
    df_city = df[df.place==city]
    for year in df_city.year.unique():
        blooming_day = df[(df.place==city)&(df.year==year)].blooming_day.unique()[0]
        d[city][year] = blooming_day
```

```python
# -------------------------------------------------------------------------------- NOTEBOOK-CELL: CODE
d
```

```python
# -------------------------------------------------------------------------------- NOTEBOOK-CELL: CODE
df_bloom = df[df.flower_status=='bloom']
df_bloom
```

```python
# -------------------------------------------------------------------------------- NOTEBOOK-CELL: CODE
df_new = pd.DataFrame()
for city in df.place.unique():
    print(city)
    df_sub = df[df['place']==city]
    year_seq = df_sub[df_sub['date_parsed']>df_sub['blooming_day']].year.values
    shifted_year_seq = [d[city][year+1] if year < 2021 else np.nan for year in year_seq]
    df_sub.loc[df_sub['date_parsed']>df_sub['blooming_day'], "blooming_day"] = shifted_year_seq
    df_new = pd.concat([df_new, df_sub])
```

```python
# -------------------------------------------------------------------------------- NOTEBOOK-CELL: CODE
df_new[df_new['place']=='Hirosaki'].tail(365)
```

```python
# -------------------------------------------------------------------------------- NOTEBOOK-CELL: CODE
# Write recipe outputs
blooming_day_shift = dataiku.Dataset("blooming_day_shift")
blooming_day_shift.write_with_schema(df_new)
```
