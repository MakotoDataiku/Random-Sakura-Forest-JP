```python
import dataiku
from dataiku import pandasutils as pdu
import pandas as pd
```

```python
# Example: load a DSS dataset as a Pandas dataframe
mydataset = dataiku.Dataset("evaluated_joiend_prep")
df = mydataset.get_dataframe()
```

```python
df.head()
```

```python
from sklearn.metrics import mean_squared_error 
realVals = df[u'開花日までの日数']
predictedVals = df[u'開花日までの日数_predicted']
mse = mean_squared_error(realVals, predictedVals)
mse
```

```python
rmse = mean_squared_error(realVals, predictedVals, squared = False)
```
